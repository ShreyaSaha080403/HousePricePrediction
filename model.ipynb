```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
train_data = pd.read_csv('beta.csv')
```

```python
train_data.info()
```

```python
train_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
OHE = OneHotEncoder()
```

```python
OHE.fit(train_data[['powerStatusD', 'furnishedD', 'transactionTypeD', 'propTypeD', 'waterStatus', 'isPrimeLocProp']])
```

```python
train_encoded = OHE.transform(train_data[['powerStatusD', 'furnishedD', 'transactionTypeD', 'propTypeD', 'waterStatus', 'isPrimeLocProp']])
```

```python
train_encoded
```

```python
train_encoded = pd.DataFrame(train_encoded.toarray(), columns= OHE.get_feature_names_out(['powerStatusD', 'furnishedD', 'transactionTypeD', 'propTypeD', 'waterStatus', 'isPrimeLocProp']))
```

```python
train_encoded
```

```python
train_data = pd.concat([train_data, train_encoded], axis=1)
```

```python
train_data
```

```python
train_data.drop(['powerStatusD', 'furnishedD', 'transactionTypeD', 'propTypeD', 'waterStatus', 'isPrimeLocProp'], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
import pickle

with open('onehotencoder.pkl', 'wb') as file:
    pickle.dump(OHE, file)
```

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
xgb = XGBRegressor(booster='gbtree',
                   eval_metric='rmse',
                   early_stopping_rounds=100)

lgbm = LGBMRegressor(learning_rate=0.025, 
                    n_estimators=100_000, 
                    metric='rmse')
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error

kf = KFold(shuffle=True, random_state=42)

X = train_data.drop('price', axis = 1)
y = train_data['price']

train_indices = []
valid_indices = []


for i, (train_index, valid_index) in enumerate(kf.split(X)):
    X_train, y_train = X.iloc[train_index], y.iloc[train_index]
    X_valid, y_valid = X.iloc[valid_index], y.iloc[valid_index]
        
    xgb.fit(X_train, y_train,
            eval_set = [(X_valid, y_valid)],
            verbose = False)
    
    y_pred = xgb.predict(X_valid)

    print(f"fold: {i}")
    print(f"MAE: {mean_absolute_error(y_valid, y_pred)}")
    print(f"MSE: {mean_squared_error(y_valid, y_pred)}")
    print("|||" * 10)

    train_indices.append(train_index)
    valid_indices.append(valid_index)

```

```python
final_train_index = train_indices[1]
final_valid_index = valid_indices[1]
```

```python
X_final, y_final = X.iloc[final_train_index], y.iloc[final_train_index]
X_eval, y_eval = X.iloc[final_valid_index], y.iloc[final_valid_index]
xgb.fit(X_final, y_final,
        eval_set = [(X_eval, y_eval)],
        verbose = False)

with open('xgbreg.pkl', 'wb') as file:
    pickle.dump(xgb, file)
```

```python
from lightgbm import early_stopping

kf = KFold(shuffle=True, random_state=42)

X = train_data.drop('price', axis = 1)
y = train_data['price']

train_indices = []
valid_indices = []


for i, (train_index, valid_index) in enumerate(kf.split(X)):
    X_train, y_train = X.iloc[train_index], y.iloc[train_index]
    X_valid, y_valid = X.iloc[valid_index], y.iloc[valid_index]
        
    lgbm.fit(X_train, y_train,
            eval_set = [(X_valid, y_valid)],
            callbacks=[early_stopping(500, verbose=True)])
    
    y_pred = lgbm.predict(X_valid)

    print(f"fold: {i}")
    print(f"MAE: {mean_absolute_error(y_valid, y_pred)}")
    print(f"MSE: {mean_squared_error(y_valid, y_pred)}")
    print("|||" * 10)

    train_indices.append(train_index)
    valid_indices.append(valid_index)
```
